```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating the values

x = np.random.uniform(1,20,1000)
y = np.random.uniform(0,10,1000)
```

```python
n = len(x)

w = 0
b = 0
learning_rate = 0.001
iterations = 1000

losses = []
for i in range(iterations):
  #pred
  y_pred = w*x+b

  error = y_pred - y

  #loss function
  loss = (error**2).mean()/2
  losses.append(loss)

  #gradient
  dw = (2/n)*np.sum(error*x)
  db = (2/n)*np.sum(error)

  #updtae
  w = w - learning_rate*dw
  b = b - learning_rate*db
```

```python
plt.plot(losses)
```
