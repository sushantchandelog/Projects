```python
#importing the lib
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('carbon_emmition.csv')
df.head()
```

```python
#checking the column names and need columns to keep
df.columns
```

```python
df = df[['country','year', 'population','gdp', 'co2','co2_per_capita','co2_per_gdp']].copy()
```

```python
df
```

```python
#check for the max and min year 
yearCheck=df['year'].min(), df['year'].max()
print(yearCheck)
```

```python
#checking for the Null and NAN values 
group_by_year = df.groupby('year')['gdp'].apply(lambda x: x.isnull().sum())
group_by_year.head(20)
group_by_year.tail(20)
```

```python
df_clean = df[df['year'] >= 1960]
```

```python
df_clean
```

```python
df_clean['year'].min()
```

```python
df_clean.groupby('year')['co2'].apply(lambda x: x.isnull().sum())
```

```python
#checking only the nan values
nan_values = df_clean[df_clean['gdp'].isnull()]
nan_values
```

```python
df_clean['country'].unique()
```

```python
#removing the unwanted countries
non_countries = [
    'World', 'Asia', 'Africa', 'Europe', 'North America', 'South America', 'Oceania',
    'High-income countries', 'Upper-middle income countries',
    'Lower-middle income countries', 'Low-income countries',
    'European Union (27)', 'European Union (28)', 'Asia (excl. China & India)',
    'International transport','Europe (excl. EU-27)', 'Europe (GCP)', 'Europe (excl. EU-28)'
    'International aviation', 'International shipping', 'Middle East', 'Central and South-Eastern Europe',
    'Central and Southern Asia', 'Eastern Asia', 'Latin America and the Caribbean',]

#inverting the logic "~" using this

df_clean = df_clean[~df_clean['country'].isin(non_countries)]
df_clean['country'].unique()
```

```python
#checking for the null values
df_clean.isnull().sum()
```

```python
#creating and cleaning a nun values by interpolate and flag column

df_clean = df_clean.copy()
df_clean['nan_gdp'] = df_clean['gdp'].isnull() #setting the flag column for the gdp

cols_to_interpolate = ['gdp', 'co2', 'co2_per_capita', 'co2_per_gdp','population']
for col in cols_to_interpolate:
    df_clean[col] = df_clean.groupby('country')[col].transform(lambda x: x.interpolate())
df_clean = df_clean.dropna() #dropping the remaining null values

```

```python
df_clean.drop(columns=['nan_gdp'], inplace=True)
```
